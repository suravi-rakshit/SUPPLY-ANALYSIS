```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data)
print(df.head())
grouped_by=df.groupby("Drivers Active Per Hour")["Riders Active Per Hour"].sum()
print(grouped_by)
plt.scatter(grouped_by.index,grouped_by.values)
plt.xlabel("Drivers Active Per Hour",fontsize=20)
plt.ylabel("Riders Active Per Hour",fontsize=20)
plt.title("SUPPLY ANALYSIS",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("SAYANI.X.xlsx")
df=pd.DataFrame(data)
print(df.head())
plt.hist(df["Rides Completed"],bins=20)
plt.xlabel("Rides Completed",fontsize=20)
plt.title("SUPPLY ANALYSIS",fontsize=20)
plt.show()
```
